# Text Preprocessing in NLP

## What is Text Preprocessing?

**Text Preprocessing** is the process of converting raw and unstructured text into a clean, standardized, and machine-readable format before it is used for training an NLP model.

Since machine learning algorithms cannot understand raw human language directly, text preprocessing helps remove unnecessary information while preserving meaningful content.

> **In simple words:**  
> Text Preprocessing is the process of **cleaning and preparing text data** so that computers can understand and analyze it effectively.

---

# Why is Text Preprocessing Important?

Raw text often contains:

- Uppercase and lowercase letters
- HTML tags
- URLs
- Punctuation marks
- Emojis
- Slang words
- Stop words
- Misspellings
- Extra spaces

These unnecessary elements can reduce the performance of an NLP model.

Text preprocessing helps by:

- Improving model accuracy
- Reducing noise
- Making text consistent
- Reducing computational complexity
- Improving feature extraction

---

# Example

### Original Text

```
OMG!!! 😍 I Looooove this Phone!!! Visit https://amazon.in <br>
```

### After Text Preprocessing

```
love phone
```

# Common Steps in Text Preprocessing

```
Raw Text
    ↓
Lowercasing
    ↓
Remove HTML Tags
    ↓
Remove URLs
    ↓
Remove Punctuation
    ↓
Chat Word Treatment
    ↓
Spelling Correction
    ↓
Remove Stop Words
    ↓
Handle Emojis
    ↓
Tokenization
    ↓
Stemming / Lemmatization
    ↓
Clean Text
```

# 1. Lowercasing

## What is Lowercasing?

Lowercasing converts all uppercase letters into lowercase letters.

This ensures that words with different cases are treated as the same word.

---

## Example

Before

```
Machine Learning is AWESOME
```

After

```
machine learning is awesome
```

---

## Why is Lowercasing Needed?

Without lowercasing,

```
Apple
apple
APPLE
```

would be treated as three different words.

After lowercasing,

```
apple
```

becomes the single standardized form.

---

## Advantages

- Reduces vocabulary size
- Improves consistency
- Easier text processing
- Better feature extraction

---

## Disadvantages

- May lose information where capitalization is meaningful.

Example:

```
US (United States)

us (pronoun)
```

---

# 2. Remove HTML Tags

## What are HTML Tags?

HTML tags are used to design webpages but they have no useful meaning for most NLP tasks.

Examples:

- `<html>`
- `<body>`
- `<br>`
- `<p>`
- `<div>`

---

## Example

Before

```html
<p>This phone is amazing!</p>
```

After

```
This phone is amazing!
```

---

## Why Remove HTML Tags?

HTML tags increase noise and do not contribute to understanding the actual text.

---

## Advantages

- Cleaner text
- Removes unnecessary webpage formatting
- Improves model performance

---

## Disadvantages

- Sometimes formatting information may be useful for specific tasks.

---

# 3. Remove URLs

## What is URL Removal?

URLs usually do not contribute to the meaning of a sentence.

Examples:

```
https://google.com

https://amazon.in

www.youtube.com
```

---

## Example

Before

```
Visit https://amazon.in for offers.
```

After

```
Visit for offers.
```

---

## Why Remove URLs?

URLs:

- Increase vocabulary unnecessarily
- Add noise
- Usually have little semantic meaning

---

## Advantages

- Cleaner dataset
- Smaller vocabulary
- Better feature extraction

---

## Disadvantages

- URLs may contain useful information in certain applications like phishing detection.

---

# 4. Remove Punctuation

## What is Punctuation Removal?

Punctuation marks are symbols used in writing.

Examples:

```
.
,
!
?
:
;
"
'
()
[]
```

---

## Example

Before

```
Wow!!! This phone is amazing!!!
```

After

```
Wow This phone is amazing
```

---

## Why Remove Punctuation?

Punctuation usually does not contribute to word meaning in many NLP tasks.

---

## Advantages

- Reduces noise
- Simplifies text
- Reduces vocabulary size

---

## Disadvantages

In some tasks, punctuation carries meaning.

Example:

```
Let's eat, Grandma.

Let's eat Grandma.
```

---

# 5. Chat Word Treatment

## What is Chat Word Treatment?

People often use abbreviations while chatting.

These abbreviations should be converted into their full forms.

---

## Examples

| Chat Word | Full Form |
|-----------|-----------|
| LOL | Laugh Out Loud |
| OMG | Oh My God |
| BTW | By The Way |
| BRB | Be Right Back |
| ASAP | As Soon As Possible |
| IDK | I Don't Know |
| TTYL | Talk To You Later |
| THX | Thanks |

---

## Example

Before

```
OMG! LOL 😂
```

After

```
Oh My God Laugh Out Loud
```

---

## Advantages

- Improves text understanding
- Makes slang meaningful
- Better semantic representation

---

## Disadvantages

- Requires a dictionary of chat abbreviations.
- New slang appears frequently.

---

# 6. Spelling Correction

## What is Spelling Correction?

Corrects misspelled words into their proper dictionary forms.

---

## Example

Before

```
Machien Lerning is awsome.
```

After

```
Machine Learning is awesome.
```

---

## Why is it Needed?

Misspelled words create multiple unnecessary vocabulary entries.

Example

```
Computer

Computar

Compter
```

The model may incorrectly treat them as different words.

---

## Advantages

- Improves vocabulary consistency
- Better model accuracy
- Cleaner text

---

## Disadvantages

- May incorrectly change uncommon words or names.
- Computationally expensive for large datasets.

---

# 7. Removing Stop Words

## What are Stop Words?

Stop words are very common words that usually do not carry much meaning.

Examples:

```
is

am

are

the

a

an

of

to

in

on

for
```

---

## Example

Before

```
This is a very good book.
```

After

```
good book
```

---

## Why Remove Stop Words?

Removing stop words:

- Reduces unnecessary words
- Reduces vocabulary size
- Speeds up processing

---

## Advantages

- Faster training
- Less memory usage
- Better feature extraction

---

## Disadvantages

Sometimes stop words are important.

Example

```
I do not like this movie.
```

Removing "not"

```
I do like this movie.
```

The meaning changes completely.

---

# 8. Handling Emojis

## What is Emoji Handling?

Emojis often express emotions and sentiments.

Instead of removing them, they can be converted into meaningful words.

---

## Examples

| Emoji | Meaning |
|--------|----------|
| 😊 | happy |
| 😂 | laughing |
| 😍 | love |
| 😢 | sad |
| 😡 | angry |
| 👍 | thumbs up |
| ❤️ | love |

---

## Example

Before

```
I love this movie 😍
```

After

```
I love this movie love
```

or

```
I love this movie smiling face with heart eyes
```

---

## Advantages

- Preserves sentiment
- Improves sentiment analysis
- Better emotional understanding

---

## Disadvantages

- Some emojis have multiple meanings depending on context.

---

# 9. Tokenization

## What is Tokenization?

Tokenization is the process of splitting text into smaller units called **tokens**.

A token may be:

- Word
- Sentence
- Character
- Subword

---

## Example

Sentence

```
I love NLP.
```

Word Tokens

```
["I", "love", "NLP"]
```

---

## Types of Tokenization

### Word Tokenization

```
I love NLP

↓

I
love
NLP
```

---

### Sentence Tokenization

```
Hello.
How are you?

↓

Sentence 1

Sentence 2
```

---

### Character Tokenization

```
NLP

↓

N

L

P
```

---

## Advantages

- First step of NLP
- Easy feature extraction
- Required by almost all NLP models

---

## Disadvantages

- Different languages require different tokenization rules.
- Compound words can be difficult to tokenize correctly.

---

# 10. Stemming

## What is Stemming?

Stemming reduces words to their **root form** by removing prefixes or suffixes.

The resulting word may **not be a valid dictionary word**.

---

## Examples

| Original Word | Stem |
|--------------|------|
| Playing | Play |
| Played | Play |
| Player | Player |
| Studies | Studi |
| Running | Run |

Notice:

```
Studies

↓

Studi
```

"Studi" is not a valid English word.

---

## Advantages

- Fast
- Simple
- Reduces vocabulary size

---

## Disadvantages

- Produces incomplete words
- Less accurate than lemmatization

---

# 11. Lemmatization

## What is Lemmatization?

Lemmatization converts words into their **base (dictionary) form**, known as the **lemma**.

Unlike stemming, the output is always a meaningful word.

---

## Examples

| Original Word | Lemma |
|--------------|-------|
| Running | Run |
| Better | Good |
| Studies | Study |
| Mice | Mouse |
| Children | Child |

---

## Difference Between Stemming and Lemmatization

| Word | Stemming | Lemmatization |
|------|-----------|---------------|
| Studies | Studi | Study |
| Better | Better | Good |
| Running | Run | Run |
| Mice | Mice | Mouse |

---

## Advantages

- Produces meaningful words
- More accurate
- Better semantic understanding

---

## Disadvantages

- Slower than stemming
- Requires a dictionary and grammar rules

---

# Stemming vs Lemmatization

| Feature | Stemming | Lemmatization |
|----------|-----------|---------------|
| Output | Root form | Dictionary form |
| Meaningful Words | Not Always | Yes |
| Speed | Fast | Slower |
| Accuracy | Lower | Higher |
| Uses Dictionary | No | Yes |
| Preferred For | Simple NLP tasks | Modern NLP applications |

# Complete Text Preprocessing Pipeline

```
Raw Text
    ↓
Lowercasing
    ↓
Remove HTML Tags
    ↓
Remove URLs
    ↓
Remove Punctuation
    ↓
Chat Word Treatment
    ↓
Spelling Correction
    ↓
Remove Stop Words
    ↓
Handle Emojis
    ↓
Tokenization
    ↓
Stemming / Lemmatization
    ↓
Clean Text
```

---

# Key Points to Remember

- **Text Preprocessing** cleans raw text before applying NLP techniques.
- **Lowercasing** ensures consistent word representation.
- **HTML Tags** and **URLs** are usually removed because they add noise.
- **Punctuation Removal** simplifies text but should be used carefully when punctuation affects meaning.
- **Chat Word Treatment** expands abbreviations like "LOL" and "OMG" into meaningful text.
- **Spelling Correction** improves vocabulary consistency.
- **Stop Word Removal** reduces unnecessary words but should preserve important words like "not" when needed.
- **Emoji Handling** preserves emotional information, which is valuable in sentiment analysis.
- **Tokenization** splits text into words, sentences, characters, or subwords.
- **Stemming** produces root forms that may not be valid words, while **Lemmatization** returns meaningful dictionary words.
