# Feature Extraction from Text in NLP

## What is Feature Extraction from Text?

**Feature Extraction** is the process of converting **text data into numerical features (vectors)** so that Machine Learning and Deep Learning algorithms can understand and process it.

Computers cannot understand human language directly. They can only perform mathematical operations on numbers.

Therefore, before training an NLP model, text must be converted into numbers.

> **In simple words:**  
> Feature Extraction is the process of **converting text into numerical vectors** that can be understood by a computer.

---

# Example

Original Text

```
I love Machine Learning
```

Machine Learning algorithm sees

```
[0.23, 0.67, 0.12, 0.91]
```

instead of words.

---

# Why Do We Need Feature Extraction?

Machine Learning algorithms perform mathematical calculations like:

- Addition
- Multiplication
- Matrix Operations
- Gradient Descent

They **cannot understand words** like:

```
Apple

Dog

Computer

Love
```

Therefore,

```
Text

↓

Numbers

↓

Machine Learning Model
```

---

# Why is Feature Extraction Difficult?

Converting text into numbers is difficult because natural language is very complex.

Some challenges include:

### 1. Same Meaning, Different Words

```
Good

Excellent

Fantastic

Amazing
```

All have similar meanings but different words.

---

### 2. Same Word, Different Meaning

Example

```
Bank
```

Can mean

```
River Bank
```

or

```
Money Bank
```

---

### 3. Word Order Matters

Example

```
Dog bites man

Man bites dog
```

Both sentences contain the same words but have completely different meanings.

---

### 4. Synonyms

```
Car

Automobile

Vehicle
```

Same meaning, different words.

---

### 5. Polysemy

One word having multiple meanings.

Example

```
Bat
```

Can mean

- Animal
- Cricket Bat

---

### 6. Large Vocabulary

English contains hundreds of thousands of words.

Creating numerical representations for all words is difficult.

---

# Core Idea Behind Feature Extraction

The main idea is

```
Human Language

↓

Mathematical Representation

↓

Machine Learning
```

Every sentence is converted into a numerical vector.

Example

```
"I love AI"

↓

[0,1,0,1,1]
```

These vectors are then used to train the model.

---

# Common Terms

Before learning feature extraction techniques, we should understand some common terms.

---

## 1. Corpus

A **Corpus** is the complete collection of documents or text.

Example

```
Document 1

I love AI

Document 2

AI is amazing

Document 3

Machine Learning is fun
```

All documents together form the **Corpus**.

---

## 2. Document

A single text file, sentence, paragraph, or review.

Example

```
I love NLP.
```

This is one document.

---

## 3. Vocabulary

Vocabulary is the collection of all **unique words** in the corpus.

Example

Corpus

```
I love AI

AI is amazing
```

Vocabulary

```
I

love

AI

is

amazing
```

---

## 4. Token

A token is the smallest unit after tokenization.

Example

Sentence

```
I love NLP
```

Tokens

```
I

love

NLP
```

---

## 5. Vector

A numerical representation of text.

Example

```
[1,0,0,1,1]
```

# 1. One Hot Encoding (OHE)

## What is One Hot Encoding?

One Hot Encoding represents every word as a binary vector.

Only one position contains **1**, while all others contain **0**.

---

# Example

Vocabulary

```
Cat

Dog

Fish

Bird
```

Encoding

| Word | Vector |
|------|---------|
| Cat | [1,0,0,0] |
| Dog | [0,1,0,0] |
| Fish | [0,0,1,0] |
| Bird | [0,0,0,1] |

---

## Characteristics

- Binary representation
- Fixed vocabulary
- Each word is independent

---

## Advantages

- Very simple
- Easy to understand
- Easy to implement
- Useful for small datasets

---

## Disadvantages

- Very high dimensional vectors
- Sparse representation
- No semantic meaning
- Cannot capture similarity

Example

```
Cat

Dog
```

Both are animals but their vectors are completely different.

---

# 2. Bag of Words (BoW)

## What is Bag of Words?

Bag of Words represents a document using **word frequency**.

It counts how many times each word appears.

Word order is ignored.

---

# Example

Sentence 1

```
I love AI
```

Sentence 2

```
AI is amazing
```

Vocabulary

```
I

love

AI

is

amazing
```

Feature Matrix

| Sentence | I | Love | AI | Is | Amazing |
|----------|---|------|----|----|----------|
| I love AI |1|1|1|0|0|
| AI is amazing |0|0|1|1|1|

---

## Characteristics

- Counts word frequency
- Ignores grammar
- Ignores word order

---

## Advantages

- Very simple
- Fast
- Works well for text classification
- Easy to implement

---

## Disadvantages

- Ignores word order
- Large sparse matrix
- No semantic meaning
- Vocabulary grows quickly

---

# 3. N-Grams

## What is an N-Gram?

An **N-Gram** is a sequence of **N consecutive words**.

Instead of looking at single words,

it considers neighboring words together.

---

# Types of N-Grams

## Unigram (N = 1)

One word at a time.

Example

```
I love NLP
```

Output

```
I

love

NLP
```

---

## Bigram (N = 2)

Two consecutive words.

Example

```
I love NLP
```

Output

```
I love

love NLP
```

---

## Trigram (N = 3)

Three consecutive words.

Example

```
I love Machine Learning
```

Output

```
I love Machine

love Machine Learning
```

---

## 4-Gram

```
I love Machine Learning

↓

I love Machine Learning
```

---

# Why N-Grams?

Bag of Words ignores word order.

Example

```
I love AI

AI love I
```

BoW treats both as identical.

N-Grams preserve some word order.

---

# Advantages

- Preserves local context
- Captures phrases
- Better than Bag of Words
- Improves language modeling

---

# Disadvantages

- Very large vocabulary
- High memory usage
- Sparse matrix
- Computationally expensive

---

# 4. TF-IDF

## What is TF-IDF?

TF-IDF stands for

```
Term Frequency

×

Inverse Document Frequency
```

Instead of simply counting words,

TF-IDF gives more importance to important words.

Common words receive lower importance.

---

# Formula

## Term Frequency (TF)

$$
TF=\frac{\text{Number of times the term appears in a document}}{\text{Total number of words in that document}}
$$

---

## Inverse Document Frequency (IDF)

$$
IDF=\log\left(\frac{\text{Total Number of Documents}}{\text{Number of Documents containing the term}}\right)
$$

---

## TF-IDF

$$
TF\text{-}IDF = TF \times IDF
$$

---

# Example

Suppose

```
the
```

appears in every document.

Its IDF becomes very small.

Therefore,

```
TF-IDF

↓

Very Low
```

Now consider

```
Transformer
```

which appears in only a few documents.

Its IDF becomes high.

Therefore,

```
TF-IDF

↓

High
```

---

# Characteristics

- Gives higher weight to important words.
- Reduces importance of common words.
- Better than Bag of Words.

---

# Advantages

- Better feature representation
- Reduces effect of common words
- Widely used
- Works well for document classification
- Improves search engines

---

# Disadvantages

- Ignores semantics
- Ignores word order
- Cannot understand context
- Sparse representation

---

Custom Features are manually designed features that provide additional information beyond words.

These features depend on the specific NLP task.

---

# Examples

### Review Length

```
Number of Words
```

---

### Number of Characters

```
Sentence Length
```

---

### Number of Exclamation Marks

```
!!!

```

May indicate excitement.

---

### Number of Capital Letters

```
AMAZING
```

May indicate emphasis.

### Number of Emojis

```
😍😂😊
```

---

### Average Word Length

Longer words may indicate more formal writing.

### Number of Positive Words

Example

```
Excellent

Amazing

Fantastic
```

---

### Number of Negative Words

Example

## Advantages

- Improves model performance
- Captures useful information ignored by standard techniques
- Easy to combine with BoW or TF-IDF

---

## Disadvantages

- Requires manual feature engineering
- Time consuming
- Depends on domain knowledge

# Comparison of Feature Extraction Techniques

| Technique | Preserves Word Order | Captures Meaning | Sparse Matrix | Easy to Implement |
|-----------|----------------------|------------------|---------------|-------------------|
| One Hot Encoding | ❌ No | ❌ No | ✅ Yes | ✅ Very Easy |
| Bag of Words | ❌ No | ❌ No | ✅ Yes | ✅ Easy |
| N-Grams | ✅ Partially | ❌ No | ✅ Yes | ✅ Moderate |
| TF-IDF | ❌ No | ❌ No | ✅ Yes | ✅ Easy |
| Custom Features | Depends | Depends | No | Moderate |

---

# Which Technique is Better?

| Scenario | Recommended Technique |
|-----------|-----------------------|
| Small Dataset | Bag of Words |
| Document Classification | TF-IDF |
| Phrase Detection | N-Grams |
| Search Engine | TF-IDF |
| Sentiment Analysis | TF-IDF + Custom Features |
| Modern NLP | Word Embeddings (Word2Vec, GloVe, FastText, BERT) |

---

# Complete Feature Extraction Pipeline

```
Raw Text
      ↓
Text Preprocessing
      ↓
Tokenization
      ↓
Feature Extraction
      │
      ├── One Hot Encoding
      ├── Bag of Words
      ├── N-Grams
      ├── TF-IDF
      └── Custom Features
      ↓
Numerical Feature Matrix
      ↓
Machine Learning Model
      ↓
Prediction
```

---

# Key Points to Remember

- **Feature Extraction** converts text into numerical vectors that Machine Learning and Deep Learning models can understand.
- It is necessary because algorithms perform mathematical operations and cannot process raw text directly.
- Challenges include **synonyms, polysemy, word order, and large vocabularies**.
- **One Hot Encoding** represents each word as a unique binary vector but does not capture semantic relationships.
- **Bag of Words (BoW)** counts word frequencies but ignores grammar and word order.
- **N-Grams (Unigram, Bigram, Trigram)** preserve local word order and capture short phrases but increase vocabulary size.
- **TF-IDF** assigns higher importance to rare and informative words while reducing the influence of common words.
- **Custom Features** are manually engineered features such as review length, emoji count, punctuation count, or sentiment word counts that can improve model performance for specific tasks.
- Modern NLP models often replace these techniques with **Word Embeddings** (Word2Vec, GloVe, FastText, BERT), which capture semantic meaning and context.
