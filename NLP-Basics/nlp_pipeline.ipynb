# NLP Pipeline

Natural Language Processing (NLP) is the process of enabling computers to understand, analyze, and generate human language. Building an NLP application involves several important steps, known as the **NLP Pipeline**.

A typical NLP Pipeline consists of:

1. Data Acquisition
2. Text Preparation
3. Feature Engineering
4. Modeling
5. Evaluation
6. Deployment

---

# NLP Pipeline Overview

```
Data Acquisition
        ↓
Text Preparation
        ↓
Feature Engineering
        ↓
Model Training
        ↓
Evaluation
        ↓
Deployment
```

---

# 1. Data Acquisition

## What is Data Acquisition?

Data Acquisition is the process of collecting text data that will be used to train and test an NLP model.

Since machine learning models learn from examples, having good quality data is the most important step.

# 2. Text Preparation

## What is Text Preparation?

Raw text cannot be directly used by machine learning algorithms.

Text Preparation converts messy text into a clean and standardized format.

Example:

Original Text

```
Wow!!! This Mobile is AMAZING 😍😍
```

After cleaning

```
wow this mobile is amazing
```

---

## Common Text Preparation Steps

### 1. Lowercasing

Convert all letters to lowercase.

Example

```
I Love NLP

↓

i love nlp
```

Purpose:

Treat "Apple" and "apple" as the same word.

---

### 2. Remove Punctuation

Example

```
Hello!!!

↓

Hello
```

Removes:

- !
- ?
- .
- ,
- ;
- :

---

### 3. Remove Numbers

Example

```
I bought 3 phones

↓

I bought phones
```

Sometimes numbers are useful (e.g., age, prices), so this step depends on the application.

---

### 4. Remove Extra Spaces

Example

```
Hello      World

↓

Hello World
```

---

### 5. Tokenization

Split text into individual words.

Example

```
I love machine learning

↓

["I", "love", "machine", "learning"]
```

These individual words are called **tokens**.

---

### 6. Stop Word Removal

Stop words are very common words that usually do not add much meaning.

Examples:

- is
- am
- are
- the
- a
- an
- in
- on
- of

Example

```
This is a good movie

↓

good movie
```

---

### 7. Stemming

Stemming removes word endings to get the root form.

Example

| Original | Stem |
|----------|------|
| Playing | Play |
| Played | Play |
| Playing | Play |
| Studies | Studi |

Notice that stemming may produce incomplete words.

---

### 8. Lemmatization

Lemmatization converts words into meaningful dictionary words.

Example

| Original | Lemma |
|----------|-------|
| Running | Run |
| Better | Good |
| Studies | Study |

Lemmatization is more accurate than stemming but slightly slower.

---

### 9. Remove Special Characters

Example

```
Hello @user #AI 😊

↓

Hello AI
```

---

### 10. Spelling Correction (Optional)

Example

```
Machien Lerning

↓

Machine Learning
```

---

# Output of Text Preparation

Input

```
Wow!! I absolutely loved this Mobile Phone.
```

Output

```
love mobile phone
```

Now the text is clean and ready for feature extraction.

# 3. Feature Engineering

## What is Feature Engineering?

Machine learning algorithms cannot understand text directly.

They only understand numbers.

Feature Engineering converts text into numerical vectors.

Example

```
"I love AI"

↓

[0.12, 0.84, 0.56, 0.21]
```

Now the machine learning model can process it.

---

## Common Feature Engineering Techniques

### 1. Bag of Words (BoW)

Counts how many times each word appears.

Example

Sentences:

```
I love AI

AI is amazing
```

Vocabulary

```
I
love
AI
is
amazing
```

Vector Representation

| Sentence | I | love | AI | is | amazing |
|-----------|---|------|----|----|----------|
| I love AI | 1 | 1 | 1 | 0 | 0 |
| AI is amazing | 0 | 0 | 1 | 1 | 1 |

Simple but ignores word order and context.

---

### 2. TF-IDF (Term Frequency–Inverse Document Frequency)

TF-IDF gives higher importance to important words and lower importance to very common words.

Common words like:

- the
- is
- and

receive lower scores.

Rare but meaningful words receive higher scores.

Used in:

- Search Engines
- Document Classification
- Spam Detection

---

### 3. Word Embeddings

Instead of simple counts, words are represented as dense vectors that capture meaning.

Popular methods:

- Word2Vec
- GloVe
- FastText

Example

```
King → [0.12, 0.56, 0.91...]

Queen → [0.11, 0.58, 0.89...]
```

Similar words have similar vector representations.

---

### 4. Contextual Embeddings

Modern NLP models generate embeddings based on the surrounding words (context).

Examples:

- BERT
- RoBERTa
- GPT
- DistilBERT

Example:

```
Bank (river bank)

Bank (money bank)
```

These models produce different embeddings for each meaning based on context.

# Evaluation

Evaluation measures how well the model performs.

There are two main types:

---

# Intrinsic Evaluation

## Definition

Intrinsic Evaluation measures the model using predefined evaluation metrics on benchmark datasets.

It evaluates the model itself without considering a real-world application.

Examples:

- Accuracy
- Precision
- Recall
- F1 Score
- BLEU Score
- ROUGE Score
- Perplexity

Example

Suppose a sentiment classifier predicts:

```
Positive
Negative
Positive
```

These predictions are compared with the actual labels to calculate metrics.

---

## Advantages

- Fast
- Easy to compare models
- Standard benchmarks available

---

## Disadvantages

- May not reflect real-world usefulness.

---

# Extrinsic Evaluation

## Definition

Extrinsic Evaluation measures how much the NLP model improves the performance of a complete application.

Instead of evaluating the model alone, we evaluate the final system.

Example:

A chatbot uses an NLP model.

Instead of checking only accuracy, we measure:

- Customer satisfaction
- Task completion rate
- Response quality
- User engagement

---

## Examples

- Search Engine Performance
- Chatbot Success Rate
- Translation Quality in Production
- Customer Satisfaction
- Business Impact

---

## Advantages

- Reflects real-world performance.
- Measures practical usefulness.

---

## Disadvantages

- Time-consuming
- Expensive
- Requires real users or production testing

---

# Intrinsic vs Extrinsic Evaluation

| Feature | Intrinsic Evaluation | Extrinsic Evaluation |
|----------|----------------------|----------------------|
| Measures | Model performance | Application performance |
| Environment | Benchmark dataset | Real-world system |
| Cost | Low | High |
| Speed | Fast | Slow |
| Example | Accuracy, F1, BLEU | User satisfaction, business impact |

# 5. Deployment

## What is Deployment?

Deployment is the process of making the trained NLP model available for real users.

Instead of running only on a developer's computer, the model is hosted on a server or cloud so applications can use it.

Example:

```
User

↓

Mobile App

↓

API

↓

NLP Model

↓

Prediction

↓

User
```

---

## Deployment Steps

1. Train the model.
2. Save the trained model.
3. Build an API (e.g., Flask or FastAPI).
4. Deploy it on a cloud platform.
5. Connect it with a website or mobile app.
6. Monitor and update the model over time.

# Complete NLP Pipeline Summary

```
1. Data Acquisition
        ↓
Collect text from various sources

2. Text Preparation
        ↓
Clean and preprocess text

3. Feature Engineering
        ↓
Convert text into numerical vectors

4. Modeling
        ↓
Train machine learning or deep learning models

5. Evaluation
        ↓
Measure model performance
(Intrinsic & Extrinsic)

6. Deployment
        ↓
Deploy the trained model for real-world use
```

---

# Key Points to Remember

- **Data Acquisition:** Collect relevant and high-quality text data.
- **Text Preparation:** Clean and standardize raw text.
- **Feature Engineering:** Convert text into numerical features using techniques like BoW, TF-IDF, and embeddings.
- **Modeling:** Train algorithms such as Naive Bayes, SVM, LSTM, or Transformers.
- **Intrinsic Evaluation:** Evaluate the model using standard metrics (Accuracy, F1, BLEU, ROUGE, etc.).
- **Extrinsic Evaluation:** Evaluate the impact of the model within a complete application.
- **Deployment:** Make the trained model available through APIs or cloud services for real-world users.
