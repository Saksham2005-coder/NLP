# Word Embeddings in NLP

## What are Word Embeddings?

**Word Embeddings** are a feature extraction technique that converts words into **dense numerical vectors** while preserving their **semantic (meaning) and syntactic (grammatical) relationships**.

Unlike One-Hot Encoding or Bag of Words, Word Embeddings place **similar words close to each other in the vector space**.

> **In simple words:**  
> Word Embeddings convert words into vectors such that **words with similar meanings have similar vector representations.**

---

# Why Do We Need Word Embeddings?

Earlier techniques like:

- One Hot Encoding
- Bag of Words
- TF-IDF

have several limitations.

Example

```
King

Queen

Man

Woman
```

Using One-Hot Encoding

```
King   → [1,0,0,0]

Queen  → [0,1,0,0]
```

The computer thinks both words are completely unrelated.

But in reality,

```
King

↓

Queen
```

are closely related.

Word Embeddings solve this problem.

---

# Example

Word Embeddings may represent words like this:

| Word | Vector (Example) |
|------|----------------|
| King | [0.12, 0.45, -0.23, ...] |
| Queen | [0.11, 0.47, -0.21, ...] |
| Prince | [0.13, 0.43, -0.20, ...] |
| Apple | [-0.72, 0.18, 0.95, ...] |

Notice that:

- King, Queen, Prince are close together.
- Apple is far away.

---

# Characteristics of Word Embeddings

- Dense vectors
- Lower dimensions (50, 100, 300, etc.)
- Capture semantic meaning
- Capture grammatical relationships
- Learn automatically from data

---

# Why are Word Embeddings Better?

Consider two words:

```
Car

Automobile
```

One-Hot Encoding

```
Car

↓

[1,0,0,0]

Automobile

↓

[0,1,0,0]
```

They appear completely different.

Word Embeddings

```
Car

↓

[0.45,0.23,0.78]

Automobile

↓

[0.44,0.25,0.79]
```

Almost identical vectors.

The model understands they have similar meanings.

---

# Advantages of Word Embeddings

- Captures word meaning
- Dense representation
- Smaller vector size
- Better model performance
- Learns semantic relationships

---

# Disadvantages

- Requires training data
- Cannot fully capture context (basic Word2Vec)
- Different meanings of the same word share one vector

Example

```
Bank
```

River Bank

Money Bank

Both receive the same embedding in Word2Vec.

---

# What is Word2Vec?

## Definition

**Word2Vec** is one of the most popular algorithms for learning Word Embeddings.

It was introduced by **Google in 2013**.

Instead of manually assigning vectors,

Word2Vec **learns word vectors automatically by analyzing the surrounding words (context).**

> **In simple words:**  
> Word2Vec learns the meaning of a word by observing the words that appear around it.

---

# Core Idea Behind Word2Vec

Words that appear in **similar contexts usually have similar meanings.**

This idea is based on the **Distributional Hypothesis**.

Example

Sentence 1

```
The cat drinks milk.
```

Sentence 2

```
The dog drinks milk.
```

The words

```
Cat

Dog
```

appear in similar contexts.

Word2Vec learns that they are related.

---

# Word2Vec Intuition

Imagine reading thousands of books.

You notice:

```
King

↓

Queen

↓

Prince

↓

Princess
```

often appear together.

Similarly,

```
Apple

Banana

Mango
```

appear together.

The model learns:

```
King

↓

Royal Family

Apple

↓

Fruit
```

It understands relationships without being explicitly programmed.

---

# Example

Sentence

```
I love Machine Learning.
```

Word2Vec looks at neighboring words.

```
I

↓

love

↓

Machine

↓

Learning
```

By observing these relationships across millions of sentences,

it learns meaningful word vectors.

---

# Famous Example

One of the most famous properties of Word2Vec is:

```
King

-

Man

+

Woman

≈

Queen
```

This happens because the vectors capture semantic relationships.

---

# How Word2Vec Works

Suppose we have:

```
The cat drinks milk.
```

Word2Vec selects one word.

Example

```
Cat
```

Then it looks at neighboring words.

```
The

Cat

Drinks
```

From these neighboring words,

the neural network learns the vector representation of "Cat".

---

# Types of Word2Vec Architectures

Word2Vec has two architectures:

```
Word2Vec

│

├── Continuous Bag of Words (CBOW)

└── Skip-Gram
```

---

# 1. Continuous Bag of Words (CBOW)

## What is CBOW?

CBOW predicts the **target (center) word** using the surrounding context words.

> **Context Words → Target Word**

---

# Example

Sentence

```
I love Machine Learning
```

Suppose the target word is

```
Machine
```

Context words become

```
I

love

Learning
```

CBOW learns

```
Input

↓

I

love

Learning

↓

Neural Network

↓

Predict

↓

Machine
```

---

# Another Example

Sentence

```
The cat drinks milk
```

Target Word

```
cat
```

Input

```
The

drinks
```

Output

```
cat
```

---

# CBOW Architecture

```
Context Words

↓

Embedding Layer

↓

Average Vectors

↓

Hidden Layer

↓

Softmax

↓

Target Word
```

---

# Characteristics

- Predicts center word
- Uses surrounding words
- Faster training
- Works well on large datasets

---

# Advantages

- Fast training
- Computationally efficient
- Better for frequent words
- Suitable for large datasets

---

# Disadvantages

- Less effective for rare words
- Averages context, so some information may be lost

---

# 2. Skip-Gram

## What is Skip-Gram?

Skip-Gram does the opposite of CBOW.

It predicts the **surrounding context words** using the target (center) word.

> **Target Word → Context Words**

---

# Example

Sentence

```
I love Machine Learning
```

Target Word

```
Machine
```

Output

```
I

love

Learning
```

The network learns

```
Machine

↓

Neural Network

↓

Predict

↓

I

love

Learning
```

---

# Another Example

Sentence

```
The cat drinks milk
```

Input

```
cat
```

Outputs

```
The

drinks
```

---

# Skip-Gram Architecture

```
Target Word

↓

Embedding Layer

↓

Hidden Layer

↓

Softmax

↓

Context Words
```

---

# Characteristics

- Predicts neighboring words
- Learns richer embeddings
- Better for rare words
- Slightly slower than CBOW

---

# Advantages

- Better word representations
- Works well for rare words
- Higher accuracy
- Captures semantic relationships more effectively

---

# Disadvantages

- Slower training
- More computationally expensive
- Requires more training time

---

# CBOW vs Skip-Gram

| Feature | CBOW | Skip-Gram |
|----------|------|-----------|
| Input | Context Words | Target Word |
| Output | Target Word | Context Words |
| Training Speed | Faster | Slower |
| Accuracy | Slightly Lower | Higher |
| Rare Words | Not Very Good | Very Good |
| Frequent Words | Excellent | Good |
| Computational Cost | Low | Higher |

# When to Use CBOW?

Use CBOW when:

- Dataset is large
- Faster training is required
- Most words are common
- Limited computational resources

---

# When to Use Skip-Gram?

Use Skip-Gram when:

- Dataset contains many rare words
- Better embeddings are required
- Accuracy is more important than speed
- More computational power is available

---

# Complete Word2Vec Workflow

```
Corpus
      ↓
Tokenization
      ↓
Choose Architecture
      │
      ├── CBOW
      └── Skip-Gram
      ↓
Train Neural Network
      ↓
Learn Word Embeddings
      ↓
Dense Word Vectors
      ↓
Use in NLP Models
```

# Key Points to Remember

- **Word Embeddings** convert words into **dense numerical vectors** that preserve semantic and syntactic relationships.
- **Word2Vec** is a neural network-based algorithm introduced by **Google (2013)** to automatically learn word embeddings from text.
- The main intuition behind Word2Vec is that **words appearing in similar contexts usually have similar meanings**.
- **CBOW (Continuous Bag of Words)** predicts the **target word from surrounding context words**. It is **faster** and works well for **frequent words**.
- **Skip-Gram** predicts the **surrounding context words from the target word**. It is **slower** but generally learns **better embeddings**, especially for **rare words**.
- Word2Vec produces **dense, low-dimensional vectors** that are much more informative than One-Hot Encoding, Bag of Words, or TF-IDF.
- A limitation of Word2Vec is that it assigns **one vector per word**, so it cannot distinguish between different meanings of the same word (e.g., **bank** as a financial institution vs. a river bank). Modern models like **BERT** overcome this limitation using contextual embeddings.
