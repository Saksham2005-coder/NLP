# Text Classification in NLP

## What is Text Classification?

**Text Classification** is an NLP task in which a machine learning or deep learning model automatically assigns one or more predefined categories (labels) to a given text.

The model learns patterns from previously labeled text and predicts the correct category for new unseen text.

> **In simple words:**  
> Text Classification is the process of **reading text and deciding which category it belongs to.**

---

# Example

Input

```
I absolutely loved this movie.
```

Output

```
Positive
```

---

Input

```
The service was terrible.
```

Output

```
Negative
```

---

```
Spam
```

Millions of text documents are generated every day.

Examples:

- Emails
- WhatsApp Messages
- News Articles
- Product Reviews
- Tweets
- Medical Reports

Reading them manually is impossible.

Text Classification automates this process.

---

# Types of Text Classification

Text classification can be categorized into four main types.

---

# 1. Binary Classification

## Definition

The model predicts **one of two classes**.

Example

```
Positive

Negative
```

---

Examples

- Spam / Not Spam
- Fake / Real News
- Yes / No
- Fraud / Not Fraud

---

# 2. Multi-Class Classification

## Definition

The model predicts **one class from many possible classes**.

Example

News Categories

```
Sports

Politics

Business

Technology

Entertainment
```

Only one class is selected.

---

# 3. Multi-Label Classification

## Definition

A document can belong to **multiple classes simultaneously**.

Example

Movie

```
Action

Comedy

Adventure
```

The movie belongs to all three categories.

---

# 4. Hierarchical Classification

## Definition

Classes are organized in a hierarchy.

Example

```
Electronics

↓

Mobile

↓

Android
```

Each prediction follows a tree-like structure.

# Approaches for Text Classification

There are several approaches.

```
Text Classification

│

├── Heuristic Approach

├── API Based Approach

├── BoW + Machine Learning

├── TF-IDF + Machine Learning

├── Word2Vec + Machine Learning / Deep Learning
```

---

# 1. Heuristic Approach

## What is Heuristic Approach?

A **Heuristic Approach** uses manually created rules instead of machine learning.

These rules are designed by domain experts.

---

# Example

Suppose we want Spam Detection.

Rules

```
IF

Message contains

"Lottery"

OR

"Win Money"

OR

"Prize"

↓

Spam
```

Example

```
Congratulations!!

You won ₹10 lakh.
```

↓

Spam

---

# Advantages

- Very simple
- No training data required
- Easy to implement
- Fast

---

# Disadvantages

- Poor scalability
- Difficult to maintain
- Cannot learn new patterns
- Low accuracy for complex problems

---

# 2. Using API

## What is API-Based Text Classification?

Instead of building a model from scratch,

we can use **pre-trained NLP APIs**.

The API already contains a trained model.

We simply send the text,

and it returns the predicted category.

---

# Example APIs

- OpenAI API
- Google Cloud Natural Language API
- AWS Comprehend
- Azure Text Analytics
- Hugging Face Inference API

---

# Workflow

```
Input Text

↓

Send to API

↓

Pre-trained Model

↓

Prediction
```

---

# Advantages

- Very high accuracy
- No training required
- Fast development
- Easy to integrate

---

# Disadvantages

- Internet connection required
- API cost
- Less customization
- Data privacy concerns

---

# 3. Using Bag of Words (BoW) and N-Grams

## Idea

Convert text into numerical features using:

- Bag of Words
- N-Grams

Then train a Machine Learning classifier.

---

# Pipeline

```
Raw Text

↓

Preprocessing

↓

Bag of Words

or

N-Grams

↓

Feature Matrix

↓

Machine Learning Algorithm

↓

Prediction
```

---

# Common Algorithms

- Naive Bayes
- Logistic Regression
- SVM
- Random Forest

---

# Example

Sentence

```
I love AI
```

BoW

```
[1,1,1,0,0]
```

Machine Learning Model

↓

Positive

---

# Advantages

- Simple
- Fast
- Easy to implement
- Works well for small datasets

---

# Disadvantages

- Ignores semantics
- High-dimensional vectors
- Sparse matrices
- Limited context understanding

---

# 4. Using TF-IDF

## Idea

Instead of counting words,

assign importance to each word.

Rare words receive higher importance.

Common words receive lower importance.

---

# Pipeline

```
Raw Text

↓

Preprocessing

↓

TF-IDF

↓

Feature Matrix

↓

Machine Learning Model

↓

Prediction
```

---

# Example

Word

```
the
```

Low Importance

---

Word

```
Transformer
```

High Importance

---

# Common Algorithms

- Logistic Regression
- Naive Bayes
- SVM
- Random Forest

---

# Advantages

- Better than BoW
- Reduces importance of common words
- Good accuracy
- Widely used

---

# Disadvantages

- Ignores word order
- Ignores semantics
- Sparse representation

---

# 5. Using Word2Vec

## Idea

Instead of sparse vectors,

Word2Vec converts words into **dense vectors** that capture semantic meaning.

These vectors are then used for classification.

---

# Pipeline

```
Raw Text

↓

Preprocessing

↓

Word2Vec Embeddings

↓

Dense Word Vectors

↓

Machine Learning

or

Deep Learning Model

↓

Prediction
```

---

# Example

Sentence

```
I love AI
```

Word2Vec

```
[0.23, 0.56, -0.44, ...]
```

These dense vectors preserve semantic relationships.

---

# Models Used

Machine Learning

- Logistic Regression
- SVM
- Random Forest

Deep Learning

- ANN
- CNN
- LSTM
- GRU

---

# Advantages

- Captures semantic meaning
- Dense vectors
- Better than BoW and TF-IDF
- Lower-dimensional representation

---

# Disadvantages

- Same word has only one vector (no context awareness)
- Requires training data or pre-trained embeddings
- Cannot distinguish different meanings of the same word

Example

```
Bank
```

River Bank

Money Bank

Both receive the same embedding.

---

# Comparison of Approaches

| Approach | Training Required | Semantic Understanding | Accuracy | Complexity |
|-----------|------------------|------------------------|----------|------------|
| Heuristic | No | No | Low | Very Low |
| API | No (Pre-trained) | High | Very High | Very Low |
| BoW | Yes | No | Medium | Low |
| TF-IDF | Yes | No | Medium-High | Low |
| Word2Vec | Yes / Pre-trained | Yes | High | Medium |

# Which Approach Should You Use?

| Scenario | Recommended Approach |
|-----------|----------------------|
| Very Small Project | Heuristic Rules |
| Quick Production Solution | API |
| Small Dataset | BoW |
| Medium Dataset | TF-IDF |
| Semantic Understanding Needed | Word2Vec |
| Modern NLP Applications | BERT, RoBERTa, DistilBERT (Contextual Embeddings) |

# Key Points to Remember

- **Text Classification** is the process of automatically assigning predefined categories to text.
- Common types include **Binary**, **Multi-Class**, **Multi-Label**, and **Hierarchical Classification**.
- Major applications include **Spam Detection, Sentiment Analysis, News Classification, Language Detection, Intent Detection, Toxic Comment Detection, and Medical Document Classification**.
- A typical pipeline consists of **Text Preprocessing → Tokenization → Feature Extraction → Model Training → Prediction**.
- **Heuristic Approach** uses manually designed rules and is suitable for simple tasks.
- **API-Based Classification** uses pre-trained cloud models and requires no training but may involve cost and internet dependency.
- **Bag of Words (BoW)** and **N-Grams** convert text into frequency-based vectors and work well for traditional machine learning models.
- **TF-IDF** improves upon BoW by assigning higher importance to informative words and reducing the weight of common words.
- **Word2Vec** generates dense word embeddings that capture semantic relationships, making it more powerful than BoW and TF-IDF for many NLP tasks.
- Modern NLP systems often use **contextual embeddings** such as **BERT** because they understand the meaning of a word based on its surrounding context.
